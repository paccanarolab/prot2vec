```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from FASTATool.FastaParser import FastaFile
```

```python
fasta = FastaFile('E:/prot2vec/83333/83333.fasta')
```

```python
fasta.buildBrowsableDict()
```

```python
ecoli = pd.DataFrame(fasta.information['proteins'].items(), columns=['protein', 'sequence'])
```

```python
ecoli['seqlen'] = ecoli['sequence'].apply(len)
```

```python
fig, ax = plt.subplots(figsize=(15,5))
median = ecoli['seqlen'].median()
sns.histplot(data=ecoli, x='seqlen', color='lightsteelblue', ax=ax)
ax.axvline(median, color='crimson')
ax.annotate(f' {median}', (median, 770), color='crimson', fontsize=16)
ax.set_xlabel('Sequence Length', fontsize=16)
ax.set_ylabel('Protein Count', fontsize=16)
ax.set_title('Sequence Length Distribution', color='crimson', fontsize=20, weight='bold')
plt.show()
plt.close('all')
```
